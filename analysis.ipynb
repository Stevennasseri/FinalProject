### 1.	Which department suffered the most losses in employees?

```python
print(df['Department'].value_counts(), '\n\n', df['Attrition'].value_counts())
```

```python
# Filter employees who left
left_employees = df[df['Attrition'] == 'Yes']

# Analyze losses by department
department_losses = left_employees['Department'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 6))
plot = sns.barplot(x=department_losses.index, y=department_losses.values)
plt.title('Employee Losses by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')


# Add value labels on top of each bar
for i, v in enumerate(department_losses.values):
    plot.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

##### R&D clearly suffered the most losses, but given the difference of size in each department, 
##### its safe to say the losses were relatively proportional to one another

```python
wholeDept = []
for key, value in df['Department'].value_counts().items():
    ## add value to wholeDept
    wholeDept.append(value)

leftDept = []
for key, value in department_losses.items():
    ## add value to wholeDept
    leftDept.append(value)

## calculate the percentage of employees who left
percent = []
for i in range(len(leftDept)):
    percent.append((leftDept[i] / wholeDept[i])*100)

## plot the bar chart
plt.figure(figsize=(10, 6))
plot = sns.barplot(x=percent, y=department_losses.index, color='red')
plt.title('Employee Losses by Department')
plt.xlabel('Percentage of Attrition')
plt.ylabel('Department')

# Add value labels on top of each bar
for i, v in enumerate(percent):
    plot.text(v, i, str(round(v, 2))+'%', ha='left', va='center')
```

### 2.	Did People with more years at the company have more security?

```python
## find how many people are in each year of employment
years = df['YearsAtCompany'].value_counts()

## find how many people left in each year of employment
left = df[df['Attrition'] == 'Yes']['YearsAtCompany'].value_counts()

## calculate the percentage of employees who left
percent = (left / years) * 100

## remove NaN values
percent = percent.dropna()

## convert index to string
percent.index = percent.index.astype(str)

## plot the bar chart
plt.figure(figsize=(10, 6))
plot = sns.barplot(y=percent.index, x=percent.values, color='skyblue')
plt.title('Employee Losses by Years at Company')
plt.xlabel('Percentage of Attrition')
plt.ylabel('Years at Company')

# Add value labels on top of each bar
for i, v in enumerate(percent):
    plot.text(v, i, str(round(v, 2))+'%', ha='left', va='center')
```

### 4.	What effect does a good work life balance have on employee effort?

```python
print(df['WorkLifeBalance'].value_counts(), '\n', df['JobInvolvement'].value_counts())
```

```python

count_work_life = df['WorkLifeBalance'].value_counts().to_dict()
count_job_involvement = df['JobInvolvement'].value_counts().to_dict()

# Create df's for easier plotting
df_work_life = pd.DataFrame({'Rating': list(count_work_life.keys()), 
                             'Count': list(count_work_life.values()), 
                             'Category': 'WorkLifeBalance'})

df_job_involvement = pd.DataFrame({'Rating': list(count_job_involvement.keys()), 
                                   'Count': list(count_job_involvement.values()), 
                                   'Category': 'JobInvolvement'})

# Combine DataFrames
df_combined = pd.concat([df_work_life, df_job_involvement])

# Plotting side-by-side bar plots
plt.figure(figsize=(12, 8))

plot = sns.barplot(x='Rating', y='Count', hue='Category', data=df_combined, palette=['skyblue', 'orange'])
plt.title('Distribution of WorkLifeBalance and JobInvolvement')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.legend()


## AI ASSISTANCE FOR THIS NEXT LINE ## - purpose: to add labels to the top of each bar
for p in plot.patches:
    plot.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

### 5.	Do employees who work more hours have higher performance ratings? Is this reflected in their pay / promotions / bigger bonus?

```python
print(df['PerformanceRating'].value_counts(), 
      '\n\n', df['StandardHours'].value_counts(), 
      '\n\n', df['OverTime'].value_counts(),
      '\n\n', df['PercentSalaryHike'].value_counts())
```

```python

# bar plot comparing performance ratings for employees who work overtime and those who don't
plt.figure(figsize=(8, 6))
plot = sns.countplot(x='PerformanceRating', hue='OverTime', data=df, palette='viridis')

# Set plot labels and title
plt.xlabel('Performance Rating')
plt.ylabel('Count')
plt.title('Distribution of Performance Ratings for Employees with and without Overtime')

# Show the legend
plt.legend(title='Overtime', loc='upper right')

## AI ASSISTANCE FOR THIS NEXT LINE ## - purpose: to add labels to the top of each bar
for p in plot.patches:
    plot.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

# Show the plot
plt.show()
```

```python

# Create a box plot to show the relationship between PerformanceRating and PercentSalaryHike
plt.figure(figsize=(6, 6))
sns.boxplot(x='PerformanceRating', y='PercentSalaryHike', data=df, palette='viridis')

# Set plot labels and title
plt.xlabel('Performance Rating')
plt.ylabel('Percent Salary Hike')
plt.title('Relationship between Performance Rating and Percent Salary Hike')

# place line at min value fror each performance rating
plt.axhline(df[df['PerformanceRating'] == 4]['PercentSalaryHike'].min(), color='red', linestyle='dashed', linewidth=2)
plt.axhline(df[df['PerformanceRating'] == 3]['PercentSalaryHike'].max(), color='red', linestyle='dashed', linewidth=2)
# Show the plot
plt.show()
```
